```python
import sys
sys.path.append('/home/evv/PycharmProjects/datahound/')
import community as cm
from getters import *
import pandas as pd

pb = cm.Community("darcor")
posts = get_posts("darcor",pb.posts_count,debug=True)
```

```python
likes = []
reposts = []
text = []
views = []
for post in posts:
    likes.append(post['likes']['count'])
    reposts.append(post['reposts']['count'])
    text.append(post['text'])
    views_count = post.get('views',None)
    if views_count is not None:
        views.append(views_count.get('count',None))
    else:
        views.append(views_count)
    
print(len(views) == len(text) and len(reposts) == len(likes))
```

```python
import re
tags = []
for each in text:
    tags.append(re.findall(r'\#\w+', each))
```
